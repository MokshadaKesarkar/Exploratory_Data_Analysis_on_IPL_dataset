1. Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Loading Dataset


```python
deliveries_data = pd.read_csv("/content/drive/MyDrive/internshiip/deliveries.csv")
```

```python
matches_data = pd.read_csv("/content/drive/MyDrive/internshiip/matches.csv")
```

3. Merging the dataset

```python
merge_df = pd.merge(deliveries_data,matches_data,left_on='match_id',right_on='id')

merge_df.head()
```

4. Information about dataset

```python
merge_df.info()
```

```python
merge_df.isna().any()
```

6. Missing values


```python
merge_df.isnull().sum()
```

```python
merge_df.describe()
```

8. Number of Matches played in each IPL Season

```python
plt.figure(figsize = (20,12))
sns.countplot('season' , data= matches_data,)
plt.xlabel("Season Played" , fontsize = 18)
plt.ylabel("Matches" , fontsize = 18)
plt.title('Matches Played in each IPL season', fontsize = 18)
plt.show()
```

From the above graph we conclude that most matches were played in season 2013 , season 2012, season 2011

9. Number of Matches won by a team 

```python
plt.figure(figsize = (18,10))
sns.countplot('winner', data=matches_data, palette = 'pastel')
plt.xlabel("Teams", fontsize = 16)
plt.xticks(rotation = 60)
plt.ylabel("Matches won", fontsize = 16)
plt.title('Matches won by a team', fontsize = 16)
plt.show()
```

From the above graph we conclude that most of the matches won by teams are Mumbai Indians, Chennai super kings, Kolkata Night Riders


10. Number of Matches Played in Stadium


```python
plt.figure(figsize = (18,10))
sns.countplot('venue', data=matches_data)
plt.xlabel("Venues", fontsize = 16)
plt.xticks(rotation = 90)
plt.ylabel("Matches played", fontsize = 16)
plt.title('Number of Matches Played in Stadium', fontsize = 16)
plt.show()
```

From the above graph we conclude that most of the matches were played in Eden Gardens

11. Toss Decision by teams whether to bat or ball

```python
plt.figure(figsize=(6,6))
sizes = matches_data.toss_decision.value_counts()
labels = matches_data.toss_decision.value_counts().index
plt.pie(sizes,colors = ['#ff9999','#66b3ff'],
         labels=labels,
         autopct='%1.1f%%',
         startangle=90,)
plt.show()
```

From the above graph we conclude that most teams decided to field 


12. Team Decision to Bat or Field

```python
plt.figure(figsize=(18,10))
sns.countplot('toss_winner' , data=matches_data , hue='toss_decision')
plt.xlabel('Teams')
plt.xticks(rotation = 90)
plt.ylabel('Counts')
plt.title('Teams Decision to bat or field after winning the toss', size = 16)
plt.show()
```

From the above graph we conclude that Kolkata Knight Riders chose to field more than to bat.

13. Player of the match





```python
plt.figure(figsize=(18,10))
player_of_match=matches_data['player_of_match'].value_counts()[:10]
sns.barplot(player_of_match.index,
            player_of_match.values,
            palette='pastel')

plt.title("Player of the Match",fontsize=15)
plt.xlabel('Player',fontsize=12)
plt.ylabel('Count',fontsize=12)
```

From the above graph we conclude that Chris Gayle has scored most of the Player of match.

14. Total Runs in IPL season

```python
merge_df.groupby('season')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'b',
                                                                                            figsize = (15,5))
plt.title("Top Runs of IPL Season", fontsize =18)
plt.xlabel("Season")
plt.ylabel("Total Runs Scored")
plt.xticks(size = 12)
plt.yticks(size = 12)
```

From the above graph we conclude that most of the runs were scored during the season 2013 and season 2012




15. Total Runs of each batsmen in IPL season


```python
merge_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'c',figsize = (18,10))
plt.xlabel('batsman',fontsize=12)
plt.ylabel('Runs scored',fontsize=12)
plt.title('Total runs scored by each batsmen in IPL Season',fontsize=16)
```

From the above graph we conclude that most runs were scored by Virat Kohli

16. Top wicket takers of IPL

```python
merge_df.groupby('bowler')['player_dismissed'].count().sort_values(ascending = False).head(10).plot(kind = 'bar', 
                                                color = 'y', figsize = (15,5))
plt.title("Top Wicket Takers of IPL", fontsize = 18)
plt.xlabel("Bowler")
plt.ylabel("Total Wickets Taken")
plt.xticks(size = 12)
plt.yticks(size = 12)
```

From the above graph we conclude that most wickets were taken by SL Malinga
